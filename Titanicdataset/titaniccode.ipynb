```python
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
sns.set()
def bar_chart(feature):
    survived = train[train['Survived']==1][feature].value_counts()
    dead = train[train['Survived']==0][feature].value_counts()
    df = pnd.DataFrame([survived,dead])
    df.index=['survived','dead']
    df.plot(kind='bar',stacked=True,figsize=(10,5))
```

```python
feature_drop = ["SibSp","Parch","Ticket"]
train = train.drop(feature_drop,axis=1)
test = test.drop(feature_drop,axis=1)
```

```python
train_data = train.drop(["PassengerId","Survived"],axis=1)
target = train["Survived"]
train_data.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import numpy as np

kfold = KFold(n_splits=10,shuffle=True,random_state=0)
```

```python
clf = KNeighborsClassifier(n_neighbors=13)
scoring = "accuracy"
score = cross_val_score(clf,train_data,target,cv=kfold,n_jobs=1,scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

```python
clf = DecisionTreeClassifier()
scoring = "accuracy"
score = cross_val_score(clf,train_data,target,cv=kfold,n_jobs=1,scoring=scoring)
print(score)
round(np.mean(score)*100,2)
```

```python
clf = RandomForestClassifier()
scoring = "accuracy"
score = cross_val_score(clf,train_data,target,cv=kfold,n_jobs=1,scoring=scoring)
print(score)
round(np.mean(score)*100,2)
```

```python
clf = GaussianNB()
scoring = "accuracy"
score = cross_val_score(clf,train_data,target,cv=kfold,n_jobs=1,scoring=scoring)
print(score)
round(np.mean(score)*100,2)
```

```python
clf = SVC()
scoring = "accuracy"
score = cross_val_score(clf,train_data,target,cv=kfold,n_jobs=1,scoring=scoring)
print(score)
round(np.mean(score)*100,2)
```

```python
clf = LogisticRegression()
scoring = "accuracy"
score = cross_val_score(clf,train_data,target,cv=kfold,n_jobs=1,scoring=scoring)
print(score)
round(np.mean(score)*100,2)
```

```python
clf = SVC()
test_data = test.drop("PassengerId",axis=1).copy()
clf.fit(train_data,target)
prediction = clf.predict(test_data)

submission = pnd.DataFrame(
    {
        "PassengerId":test["PassengerId"],
        "Survived":prediction
    }
)
submission.to_csv("submission.csv",index=False)
```
